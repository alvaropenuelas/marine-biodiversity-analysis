```python
import pandas as pd
import numpy as np

df = pd.read_csv("../data/orcinus_orca_obis.csv")
print(f"Original records: {len(df)}")
```

```python
cols = [
    "scientificName", "decimalLatitude", "decimalLongitude",
    "date_year", "depth", "country", "ocean",
    "basisOfRecord", "datasetName"
]

available = [c for c in cols if c in df.columns]
df_clean = df[available].copy()
print(f"Columns kept: {available}")
```

```python
cols = [
    "scientificName", "decimalLatitude", "decimalLongitude",
    "date_year", "depth", "country", "ocean",
    "basisOfRecord", "datasetName"
]

available = [c for c in cols if c in df.columns]
df_clean = df[available].copy()
print(f"Columns kept: {available}")
```

```python
df_clean = df_clean.dropna(subset=["decimalLatitude", "decimalLongitude"])
df_clean = df_clean.drop_duplicates(
    subset=["decimalLatitude", "decimalLongitude", "date_year"]
)
print(f"Records after removing duplicates: {len(df_clean)}")
```

```python
df_clean.to_csv("../data/orcinus_orca_clean.csv", index=False)
print(f"Clean dataset saved: {len(df_clean)} records")
print(f"\nFinal summary:")
print(df_clean.describe())
```

```python
df_clean = df_clean[
    (df_clean["date_year"] >= 1950)
]

df_clean.to_csv("../data/orcinus_orca_clean.csv", index=False)
print(f"Records after proper year filter: {len(df_clean)}")
print(f"Year range: {df_clean['date_year'].min()} - {df_clean['date_year'].max()}")
```
