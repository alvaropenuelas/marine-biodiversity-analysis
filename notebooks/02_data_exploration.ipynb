```python
cols_of_interest = [
    "scientificName", "decimalLatitude", "decimalLongitude",
    "date_year", "depth", "country", "ocean", 
    "basisOfRecord", "datasetName"
]

# Solo las que existen en tu dataset
available = [c for c in cols_of_interest if c in df.columns]
print("Available relevant columns:")
print(available)

df[available].head(10)
```

```python
print("=== TEMPORAL COVERAGE ===")
print(f"Years range: {df['date_year'].min()} - {df['date_year'].max()}")
print(f"Records with year data: {df['date_year'].notna().sum()}")

print("\n=== SPATIAL COVERAGE ===")
print(f"Latitude range: {df['decimalLatitude'].min():.2f} - {df['decimalLatitude'].max():.2f}")
print(f"Longitude range: {df['decimalLongitude'].min():.2f} - {df['decimalLongitude'].max():.2f}")
print(f"Records with coordinates: {df['decimalLatitude'].notna().sum()}")

print("\n=== DEPTH ===")
print(f"Records with depth data: {df['depth'].notna().sum()}")
```
